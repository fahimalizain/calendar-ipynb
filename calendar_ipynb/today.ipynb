```python
from datetime import datetime, time
from calendar_ipynb.events import process_events_and_classify, fetch_events_parallel, get_primary_timezone, insert_time_left_for_today
from calendar_ipynb.ipywidgets.calendar_selection import get_selection_from_cache as get_selected_calendars

%matplotlib widget

calendars = get_selected_calendars()
timezone = get_primary_timezone(calendars)
today = datetime.now().date()

from_datetime = datetime.combine(today, time.min, tzinfo=timezone)
to_datetime = datetime.combine(today, time.max, tzinfo=timezone)

# Fetch events since last 90 days
fetched_events = fetch_events_parallel(
    email_map=calendars,
    from_datetime=from_datetime,
    to_datetime=to_datetime
)

events = process_events_and_classify(
    events=fetched_events,
    from_datetime=from_datetime,
    to_datetime=datetime.now(tz=timezone),
)
events = insert_time_left_for_today(events=events, timezone=timezone)
print('Total Hrs:', sum(x['duration_min'] for x in events) / 60)
```
